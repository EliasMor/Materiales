```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos simulados (puedes reemplazarlos por los tuyos reales)
years = np.array([2019, 2020, 2021, 2022, 2023, 2024]).reshape(-1, 1)
values = np.array([89.1, 88.5, 88.3, 88.0, 81.7, 79.3])

# Ajustar modelo de regresión lineal
model = LinearRegression()
model.fit(years, values)

# Proyección hasta 2040
future_years = np.arange(2019, 2041).reshape(-1, 1)
future_values = model.predict(future_years)

# Crear DataFrame de resultados
projection_df = pd.DataFrame({
    "Año": future_years.flatten(),
    "Valor proyectado": future_values
})

# Gráfico
plt.figure(figsize=(10, 5))
plt.plot(projection_df["Año"], projection_df["Valor proyectado"], marker='o', label='Proyección')
plt.scatter(years, values, color='red', label='Datos observados')
plt.title("Proyección lineal hasta 2040")
plt.xlabel("Año")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Mostrar tabla
projection_df.tail(10)  # Mostrar últimos 10 años como ejemplo
```

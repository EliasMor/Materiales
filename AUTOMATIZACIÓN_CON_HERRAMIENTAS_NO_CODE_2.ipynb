```python
# ============================================================================
# M√ìDULO 8 - SESI√ìN 2: AUTOMATIZACI√ìN CON HERRAMIENTAS NO-CODE
# Introducci√≥n a la Automatizaci√≥n Sin C√≥digo
# ============================================================================

def crear_banner(titulo, emoji="üöÄ", ancho=80):
    """Crear banners visuales atractivos"""
    linea = "=" * ancho
    print(f"\n{linea}")
    print(f"{emoji} {titulo.center(ancho-4)} {emoji}")
    print(f"{linea}\n")

def mostrar_concepto(concepto, definicion, ejemplo="", ventajas=None):
    """Mostrar conceptos de forma clara"""
    print(f" **{concepto}**")
    print(f"   {definicion}")
    if ejemplo:
        print(f"    Ejemplo: {ejemplo}")
    if ventajas:
        print("    Ventajas:")
        for ventaja in ventajas:
            print(f"      ‚Ä¢ {ventaja}")
    print()

# ============================================================================
# PARTE 1: ¬øQU√â SON LAS HERRAMIENTAS NO-CODE?
# ============================================================================

crear_banner("INTRODUCCI√ìN A HERRAMIENTAS NO-CODE", "üéØ")

print(" **¬øQu√© significa 'No-Code'?**")
print("   No-Code significa crear automatizaciones y aplicaciones")
print("   SIN escribir c√≥digo tradicional, usando interfaces visuales")
print()

mostrar_concepto(
    "HERRAMIENTAS NO-CODE",
    "Plataformas que permiten crear automatizaciones usando interfaces gr√°ficas",
    "Zapier, Microsoft Power Automate, IFTTT, Integromat (Make)",
    ["No requiere conocimientos de programaci√≥n",
     "Configuraci√≥n r√°pida y visual",
     "Integraci√≥n f√°cil entre aplicaciones",
     "Mantenimiento simplificado"]
)

mostrar_concepto(
    "AUTOMATIZACI√ìN DE PROCESOS",
    "Configurar tareas que se ejecuten autom√°ticamente cuando ocurre un evento",
    "Cuando llega un email, guardar el adjunto en Google Drive autom√°ticamente",
    ["Ahorra tiempo manual",
     "Reduce errores humanos",
     "Funciona 24/7",
     "Escalable f√°cilmente"]
)

print(" **Flujo t√≠pico de automatizaci√≥n No-Code:**")
print("   1 **TRIGGER** (Disparador): Evento que inicia la automatizaci√≥n")
print("      ‚Ä¢ Nuevo email recibido")
print("      ‚Ä¢ Archivo subido a carpeta")
print("      ‚Ä¢ Nueva fila en Google Sheets")
print("      ‚Ä¢ Hora espec√≠fica del d√≠a")
print()
print("   2 **ACTION** (Acci√≥n): Lo que sucede autom√°ticamente")
print("      ‚Ä¢ Enviar notificaci√≥n")
print("      ‚Ä¢ Crear archivo")
print("      ‚Ä¢ Actualizar base de datos")
print("      ‚Ä¢ Generar reporte")
print()

# ============================================================================
# PARTE 2: PRINCIPALES HERRAMIENTAS NO-CODE
# ============================================================================

crear_banner("PRINCIPALES HERRAMIENTAS NO-CODE", "üõ†Ô∏è")

herramientas_nocode = {
    "Zapier": {
        "descripcion": "La plataforma m√°s popular para conectar aplicaciones web",
        "fortalezas": ["5000+ integraciones", "Interfaz muy intuitiva", "Excelente documentaci√≥n"],
        "casos_uso": ["Email marketing", "CRM automation", "Data sync"],
        "precio": "Freemium (100 tareas/mes gratis)"
    },
    "Microsoft Power Automate": {
        "descripcion": "Herramienta de Microsoft para automatizar flujos de trabajo",
        "fortalezas": ["Integraci√≥n nativa con Office 365", "AI Builder incluido", "Plantillas predefinidas"],
        "casos_uso": ["Automatizaci√≥n de Office", "Aprobaciones", "Notificaciones"],
        "precio": "Incluido en Office 365"
    },
    "Make (ex-Integromat)": {
        "descripcion": "Plataforma visual para crear automatizaciones complejas",
        "fortalezas": ["Editor visual avanzado", "L√≥gica condicional", "Manejo de errores"],
        "casos_uso": ["E-commerce automation", "Data processing", "Multi-step workflows"],
        "precio": "Freemium (1000 operaciones/mes gratis)"
    },
    "IFTTT": {
        "descripcion": "If This Then That - Automatizaciones simples",
        "fortalezas": ["Muy f√°cil de usar", "Ideal para principiantes", "Muchos dispositivos IoT"],
        "casos_uso": ["Smart home", "Social media", "Personal productivity"],
        "precio": "Freemium (3 applets gratis)"
    }
}

for herramienta, info in herramientas_nocode.items():
    print(f" **{herramienta}**")
    print(f"    {info['descripcion']}")
    print(f"    Fortalezas: {', '.join(info['fortalezas'])}")
    print(f"    Casos de uso: {', '.join(info['casos_uso'])}")
    print(f"    Precio: {info['precio']}")
    print()

# ============================================================================
# PARTE 3: COMPARACI√ìN CON AUTOMATIZACI√ìN CON C√ìDIGO
# ============================================================================

crear_banner("NO-CODE vs C√ìDIGO TRADICIONAL", "‚öñÔ∏è")

import pandas as pd

comparacion = pd.DataFrame({
    'Aspecto': [
        'Tiempo de implementaci√≥n',
        'Conocimientos requeridos',
        'Flexibilidad',
        'Costo inicial',
        'Mantenimiento',
        'Escalabilidad',
        'Control total',
        'Debugging'
    ],
    'No-Code': [
        'Minutos/Horas ',
        'B√°sicos ',
        'Media ',
        'Bajo ',
        'F√°cil ',
        'Buena ',
        'Limitado ',
        'Visual '
    ],
    'C√≥digo Tradicional': [
        'D√≠as/Semanas ',
        'Avanzados ',
        'Total ',
        'Alto ',
        'Complejo ',
        'Excelente ',
        'Completo ',
        'T√©cnico '
    ]
})

print(" **Comparaci√≥n No-Code vs C√≥digo Tradicional:**")
print(comparacion.to_string(index=False))

print("\n **¬øCu√°ndo usar cada uno?**")
print()
print(" **Usar No-Code cuando:**")
print("   ‚Ä¢ Necesitas resultados r√°pidos")
print("   ‚Ä¢ El equipo no tiene conocimientos t√©cnicos")
print("   ‚Ä¢ Son procesos est√°ndar/comunes")
print("   ‚Ä¢ El presupuesto es limitado")
print("   ‚Ä¢ Quieres prototipar r√°pidamente")
print()
print(" **Usar C√≥digo cuando:**")
print("   ‚Ä¢ Necesitas l√≥gica muy espec√≠fica")
print("   ‚Ä¢ Manejas grandes vol√∫menes de datos")
print("   ‚Ä¢ Requieres integraciones personalizadas")
print("   ‚Ä¢ La seguridad es cr√≠tica")
print("   ‚Ä¢ Necesitas control total del proceso")

# ============================================================================
# PARTE 4: CASOS DE USO REALES EN EMPRESAS
# ============================================================================

crear_banner("CASOS DE USO REALES", "üè¢")

casos_uso = [
    {
        "empresa": "Startup E-commerce",
        "problema": "Gesti√≥n manual de pedidos y inventario",
        "solucion": "Zapier conecta Shopify ‚Üí Google Sheets ‚Üí Email ‚Üí Slack",
        "beneficio": "Ahorro de 10 horas/semana, 0% errores en inventario",
        "herramientas": ["Zapier", "Shopify", "Google Sheets", "Gmail", "Slack"]
    },
    {
        "empresa": "Agencia de Marketing",
        "problema": "Reportes manuales para clientes cada semana",
        "solucion": "Power Automate genera reportes autom√°ticos desde Google Analytics",
        "beneficio": "Reportes instant√°neos, clientes m√°s satisfechos",
        "herramientas": ["Power Automate", "Google Analytics", "PowerBI", "Outlook"]
    },
    {
        "empresa": "Consultora HR",
        "problema": "Proceso de onboarding manual y lento",
        "solucion": "Make automatiza creaci√≥n de cuentas, env√≠o de documentos y seguimiento",
        "beneficio": "Onboarding de 5 d√≠as a 1 d√≠a, mejor experiencia empleado",
        "herramientas": ["Make", "Google Workspace", "DocuSign", "Slack"]
    },
    {
        "empresa": "Restaurante",
        "problema": "Gesti√≥n de reservas y recordatorios manuales",
        "solucion": "IFTTT conecta sistema de reservas con WhatsApp y Google Calendar",
        "beneficio": "95% menos no-shows, mejor organizaci√≥n",
        "herramientas": ["IFTTT", "OpenTable", "WhatsApp Business", "Google Calendar"]
    }
]

for i, caso in enumerate(casos_uso, 1):
    print(f" **Caso {i}: {caso['empresa']}**")
    print(f"    Problema: {caso['problema']}")
    print(f"    Soluci√≥n: {caso['solucion']}")
    print(f"    Beneficio: {caso['beneficio']}")
    print(f"    Herramientas: {' ‚Üí '.join(caso['herramientas'])}")
    print()

# ============================================================================
# PARTE 5: SIMULACI√ìN DE HERRAMIENTAS NO-CODE EN PYTHON
# ============================================================================

crear_banner("SIMULANDO HERRAMIENTAS NO-CODE CON PYTHON", "üêç")

print(" **Para fines educativos, vamos a simular c√≥mo funcionan**")
print("   las herramientas No-Code usando Python")
print()

class SimuladorZapier:
    """Simulador educativo de c√≥mo funciona Zapier"""

    def __init__(self):
        self.zaps = []  # Lista de automatizaciones
        self.logs = []  # Registro de ejecuciones

    def crear_zap(self, nombre, trigger, actions):
        """Crear una nueva automatizaci√≥n (Zap)"""
        zap = {
            'id': len(self.zaps) + 1,
            'nombre': nombre,
            'trigger': trigger,
            'actions': actions,
            'activo': True,
            'ejecuciones': 0
        }
        self.zaps.append(zap)
        print(f" Zap creado: '{nombre}' (ID: {zap['id']})")
        return zap['id']

    def simular_trigger(self, trigger_type, data):
        """Simular que ocurre un evento disparador"""
        print(f" Trigger detectado: {trigger_type}")

        # Buscar zaps que respondan a este trigger
        zaps_activados = [z for z in self.zaps if z['trigger']['type'] == trigger_type and z['activo']]

        for zap in zaps_activados:
            self.ejecutar_zap(zap, data)

    def ejecutar_zap(self, zap, data):
        """Ejecutar las acciones de un zap"""
        print(f" Ejecutando Zap: {zap['nombre']}")

        try:
            for i, action in enumerate(zap['actions'], 1):
                print(f"   {i}. {action['descripcion']}")

                # Simular la ejecuci√≥n de la acci√≥n
                if action['type'] == 'email':
                    self.simular_envio_email(action, data)
                elif action['type'] == 'sheets':
                    self.simular_google_sheets(action, data)
                elif action['type'] == 'slack':
                    self.simular_slack(action, data)

                # Peque√±a pausa para simular procesamiento
                import time
                time.sleep(0.5)

            zap['ejecuciones'] += 1
            self.logs.append({
                'zap_id': zap['id'],
                'timestamp': pd.Timestamp.now(),
                'estado': 'exitoso',
                'data': data
            })

            print(f" Zap '{zap['nombre']}' ejecutado exitosamente")

        except Exception as e:
            print(f" Error ejecutando Zap: {e}")
            self.logs.append({
                'zap_id': zap['id'],
                'timestamp': pd.Timestamp.now(),
                'estado': 'error',
                'error': str(e)
            })

    def simular_envio_email(self, action, data):
        """Simular env√≠o de email"""
        destinatario = action.get('destinatario', 'usuario@ejemplo.com')
        asunto = action.get('asunto', 'Notificaci√≥n autom√°tica')
        print(f"       Email enviado a {destinatario}")
        print(f"       Asunto: {asunto}")

    def simular_google_sheets(self, action, data):
        """Simular actualizaci√≥n de Google Sheets"""
        hoja = action.get('hoja', 'Hoja1')
        print(f"       Datos agregados a Google Sheets - {hoja}")
        print(f"       Datos: {data}")

    def simular_slack(self, action, data):
        """Simular mensaje de Slack"""
        canal = action.get('canal', '#general')
        mensaje = action.get('mensaje', 'Notificaci√≥n autom√°tica')
        print(f"       Mensaje enviado a Slack - {canal}")
        print(f"       Mensaje: {mensaje}")

    def mostrar_estadisticas(self):
        """Mostrar estad√≠sticas de los zaps"""
        print("\n **ESTAD√çSTICAS DE AUTOMATIZACIONES**")
        print("=" * 50)

        if not self.zaps:
            print("No hay zaps creados")
            return

        df_zaps = pd.DataFrame(self.zaps)
        print(" **Zaps activos:**")
        print(df_zaps[['id', 'nombre', 'ejecuciones', 'activo']].to_string(index=False))

        if self.logs:
            df_logs = pd.DataFrame(self.logs)
            print(f"\n **Resumen de ejecuciones:**")
            print(f"   Total ejecuciones: {len(self.logs)}")
            print(f"   Exitosas: {len(df_logs[df_logs['estado'] == 'exitoso'])}")
            print(f"   Con errores: {len(df_logs[df_logs['estado'] == 'error'])}")

# Demostraci√≥n del simulador
print(" **Creando automatizaciones con nuestro simulador:**")

zapier_sim = SimuladorZapier()

# Crear algunos zaps de ejemplo
print("\n 1 **Creando Zap: Nuevo cliente ‚Üí Bienvenida**")
zap1_id = zapier_sim.crear_zap(
    nombre="Bienvenida Nuevo Cliente",
    trigger={'type': 'nuevo_cliente', 'fuente': 'formulario_web'},
    actions=[
        {
            'type': 'email',
            'descripcion': 'Enviar email de bienvenida',
            'destinatario': 'cliente@ejemplo.com',
            'asunto': '¬°Bienvenido a nuestra empresa!'
        },
        {
            'type': 'sheets',
            'descripcion': 'Agregar cliente a Google Sheets',
            'hoja': 'Clientes'
        },
        {
            'type': 'slack',
            'descripcion': 'Notificar al equipo de ventas',
            'canal': '#ventas',
            'mensaje': 'Nuevo cliente registrado'
        }
    ]
)

print("\n 2 **Creando Zap: Venta realizada ‚Üí Reporte**")
zap2_id = zapier_sim.crear_zap(
    nombre="Reporte de Venta",
    trigger={'type': 'nueva_venta', 'fuente': 'sistema_pos'},
    actions=[
        {
            'type': 'sheets',
            'descripcion': 'Actualizar hoja de ventas',
            'hoja': 'Ventas_Diarias'
        },
        {
            'type': 'email',
            'descripcion': 'Enviar reporte al gerente',
            'destinatario': 'gerente@empresa.com',
            'asunto': 'Nueva venta registrada'
        }
    ]
)

# Simular algunos eventos
print("\n **Simulando eventos que activan los zaps:**")

print("\n Simulando: Nuevo cliente se registra...")
zapier_sim.simular_trigger('nuevo_cliente', {
    'nombre': 'Juan P√©rez',
    'email': 'juan@ejemplo.com',
    'fecha': pd.Timestamp.now()
})

print("\n Simulando: Nueva venta realizada...")
zapier_sim.simular_trigger('nueva_venta', {
    'producto': 'Laptop',
    'precio': 1200.00,
    'cliente': 'Mar√≠a Garc√≠a',
    'fecha': pd.Timestamp.now()
})

print("\n Simulando: Otro cliente se registra...")
zapier_sim.simular_trigger('nuevo_cliente', {
    'nombre': 'Ana L√≥pez',
    'email': 'ana@ejemplo.com',
    'fecha': pd.Timestamp.now()
})

# Mostrar estad√≠sticas
zapier_sim.mostrar_estadisticas()

# ============================================================================
# PARTE 6: VENTAJAS Y LIMITACIONES
# ============================================================================

crear_banner("VENTAJAS Y LIMITACIONES DEL NO-CODE", "‚öñÔ∏è")

print(" **VENTAJAS de las herramientas No-Code:**")
ventajas = [
    " Implementaci√≥n r√°pida (minutos vs d√≠as)",
    " Accesible para no-programadores",
    " Costo inicial bajo",
    " Mantenimiento simplificado",
    " Interfaces intuitivas",
    " Integraciones pre-construidas",
    " Monitoreo visual",
    " F√°cil modificaci√≥n de flujos"
]

for ventaja in ventajas:
    print(f"   {ventaja}")

print("\n **LIMITACIONES de las herramientas No-Code:**")
limitaciones = [
    " Dependencia del proveedor",
    " Costos pueden escalar r√°pidamente",
    " Limitaciones de rendimiento",
    " Funcionalidad limitada vs c√≥digo custom",
    " Menos control sobre seguridad",
    " Limitaciones en procesamiento de datos complejos",
    " Debugging m√°s dif√≠cil",
    " Escalabilidad limitada"
]

for limitacion in limitaciones:
    print(f"   {limitacion}")

print("\n **Recomendaciones para elegir:**")
print()
print(" **Usa No-Code para:**")
print("   ‚Ä¢ Automatizaciones simples a medianas")
print("   ‚Ä¢ Equipos sin conocimientos t√©cnicos")
print("   ‚Ä¢ Prototipos y MVPs")
print("   ‚Ä¢ Procesos est√°ndar de negocio")
print("   ‚Ä¢ Presupuestos limitados")
print()
print(" **Evita No-Code para:**")
print("   ‚Ä¢ L√≥gica de negocio muy compleja")
print("   ‚Ä¢ Procesamiento de big data")
print("   ‚Ä¢ Aplicaciones cr√≠ticas de seguridad")
print("   ‚Ä¢ Integraciones muy espec√≠ficas")
print("   ‚Ä¢ Cuando necesitas control total")

print("\n" + "="*80)
print(" **RESUMEN: INTRODUCCI√ìN A HERRAMIENTAS NO-CODE**")
print("="*80)
print()
print(" **Lo que hemos aprendido:**")
print("    Qu√© son las herramientas No-Code")
print("    Principales plataformas disponibles")
print("    Comparaci√≥n con desarrollo tradicional")
print("    Casos de uso reales en empresas")
print("    Simulaci√≥n de funcionamiento")
print("    Ventajas y limitaciones")
print()
print(" **Pr√≥ximo paso:**")
print("   En el siguiente notebook aprenderemos a usar Zapier")
print("   y crear nuestras primeras automatizaciones reales")
print()
print(" **Para estudiantes principiantes:**")
print("   ‚Ä¢ Enf√≥cate en entender los conceptos b√°sicos")
print("   ‚Ä¢ Piensa en procesos repetitivos en tu trabajo/estudios")
print("   ‚Ä¢ Empieza con automatizaciones simples")
print()
print(" **Para estudiantes avanzados:**")
print("   ‚Ä¢ Considera cu√°ndo usar No-Code vs c√≥digo")
print("   ‚Ä¢ Piensa en arquitecturas h√≠bridas")
print("   ‚Ä¢ Eval√∫a costos a largo plazo")
```
